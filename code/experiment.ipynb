```python
from typing import List
import math
import numpy as np
import matplotlib.pyplot as plt
import baseline_model as baseline
import elimination_model as elimination
import addition_model as addition
def plot_result(plot_name:str, consequential_region: int, positive_example_list:List[int], negative_example_list:List[int],px:List[int] ):
    positive_example_points = np.zeros(len(positive_example_list))
    negative_example_points = np.zeros(len(negative_example_list))
    plt.plot(range(1,consequential_region+1),px)
    plt.plot(positive_example_list, positive_example_points, 'bo')
    plt.plot(negative_example_list, negative_example_points, 'ro')
    plt.xlabel("y")
    plt.ylabel("p(y)")
    plt.savefig("../report/graphics/"+plot_name+".png")
```

```python

consequential_region = 120
positive_examples = [1,100]
negative_examples = [50]
```

```python
#base model, which is not able to model negative examples
px = baseline.generalize(consequential_region,positive_examples, negative_examples)
plot_result("baseline",consequential_region, positive_examples,negative_examples, px)
```

```python
# baseline model for negative examples
px2 = baseline.generalize(consequential_region,negative_examples, positive_examples)
plot_result("negative_baseline",consequential_region, positive_examples, negative_examples, px2)
```

```python
addition_model = addition.generalize(consequential_region,positive_examples, negative_examples)
plot_result("addition_model", consequential_region,positive_examples,negative_examples , addition_model)
```

```python
#elimination model
elim = elimination.generalize(consequential_region,positive_examples,negative_examples )
plot_result("elimination_model",consequential_region, positive_examples,negative_examples, elim)
```
