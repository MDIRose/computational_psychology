```python
from typing import List
import math
import numpy as np
import matplotlib.pyplot as plt
import baseline_model as baseline

def plot_result(consequential_region: int, example_list:List[int],px:List[int] ):
    example_points = np.zeros(len(example_list))
    plt.plot(range(1,consequential_region+1),px)
    plt.plot(example_list, example_points, 'ro')
    plt.xlabel("y")
    plt.ylabel("p(y)")
```

```python
consequential_region =120
example_list = [40, 50,50,60]
px = baseline.generalize(consequential_region,example_list)

plot_result(consequential_region, example_list, px)
```
